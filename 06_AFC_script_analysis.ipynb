```python

# Petite ligne de python des familles pour récupérer les "AFC FIIIILES"
def get_files_directory(directory, pattern="AFC"):
    if directory[-1] != "/":
        directory += "/"
    return [directory + fichier for fichier in os.listdir(directory) if  pattern in fichier]

def get_files_mult_directory(directory_list, pattern="AFC"):
    res = []
    for d in directory_list:
        res.extend(get_files_directory(d, pattern=pattern))
    return res
```
