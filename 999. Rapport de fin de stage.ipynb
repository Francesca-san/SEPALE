# Rapport de fin de stage:

Mon stage au sein du projet SEPALE a débuté le 25 mai 2021. Durant les premières semaines nous nous sommes concentré.e.s sur l’analyse des réponses au questionnaire soumis aux étudiant.e.s de L1, L2 et L3 de l’Université de Paris. Les réponses aux questions ont été discutées dans un rapport intitulé [Rapport à l’issue du 1er semestre du projet SEPALE](https://www.emmanuelferragne.com/project/sepale/rapportSemestre1.pdf).

<div style="text-align: justify">Dans un second temps nous avons commencé l’analyse des logfiles, fichiers créés automatiquement lorsqu’un.e étudiant.e charge un exercice sur son ordinateur. Il existe quatre types de fichiers, les 2AFC, 5AFC, AX et Oddity qui correspondent à différents exercices de discrimination de voyelles. Afin d’analyser le contenu des fichiers la première étape est de les découper, de les « parser ».</div>




<div style="text-align: center"> Figure 1 </div>


<div style="text-align: justify"> Comme la première ligne l’indique, nous avons affaire à un fichier de type AFC2 qui propose de faire la différence entre les voyelles /ᴂ/ et /ʌ/. La ligne de keys labellise les informations qui sont en colonnes, les données (= data). Enfin, un exercice fini est toujours clôturé par une dernière ligne qui correspond à ce que nous avons appelé « stats_total » et qui répertorie le nombre d’erreurs et le temps total de réalisation de l'exercice.</div>

<div style="text-align: justify"> J’ai mentionné précédemment le besoin de découper les fichiers. Ce besoin réside dans l’information que nous avons besoin de relever dans les fichiers lorsque nous appelons des fonctions. Par exemple, lorsque nous utilisons la fonction criteria_by_key, il est essentiel de faire la différence entre les clés qui nous intéressent (« Sound File », « Stimulus », « Vowel ») et les critères ( = « criteria ») que nous allons confronter afin d’obtenir des résultats (« Response Time », « Repetitions », « NbErreurs »). </div>

## III. Analyse des résultats au sein de la classe All_classrooms:

<div style="text-align: justify"> Afin d'avoir des résultats plus représentatifs nous avons créé une nouvelle classe qui regroupe tous les étudiants et étudiantes d'une classe et toutes les classes dans un même ensemble. Ceci nous permet tout d'abord de voir les résultats à l'échelle d'une classe (et plus à l'échelle de l'étudiant.e) puis, dans un second temps, de voir les résultats globaux pour tous les participants et participantes. </div>

Voici les résultats pour les L1:


<div style="text-align: center"> Figure 24: moyenne du nombre d'erreurs par voyelle (L1) </div>

<div style="text-align: center"> Figure 26: rapport entre le nombre d'erreurs et de répétitions par voyelle (L1) </div>


<div style="text-align: justify"> La première figure (Figure 24) semble montrer que la voyelle pour laquelle les étudiant.e.s ont rencontré le plus de difficultés est /ɑ:/ avec, en moyenne 0.35 erreurs. La figure suivante (Figure 25) nous permet de voir que les erreurs causées par cette voyelle surviennent en majorité lorsque /ɑ:/ est opposé à /ʌ/ puisque les stimuli qui rassemblent le plus d'erreurs sont 'card vs cud', 'buck vs bark', 'barn vs bun'. Les autres couples de termes s'appuient sur l'opposition /ᴂ/ vs /ʌ/ et /i:/ vs /ɪ/.</div>

<div style="text-align: justify"> Un autre aspect que peut être apperçu grâce à ces graphiques est la proportions d'erreurs par rapport aux répétitions. Sur la figure 26, nous pouvons voir que la voyelle /ɑ:/ pour laquelle il y a eu en moyenne le plus grand nombre d'erreurs est aussi la voyelle pour laquelle on a recencé le moins de répétitions. Au contraire, la voyelle /ᴂ/ qui ne semble pas avoir été trop problématique pour les étudiant.e.s de L1 est celle qui arrive en seconde place au niveau des répétitions, juste après /ɪ/. On peut donc percevoir un lien entre la répétition et la réussite : l'écoute attentive, et surtout la réécoute, permet une meilleure discrimination des voyelles. Ceci permet de suggérer qu'une répétitions des stimuli systématique à la fin de chaque question pourrait améliorer la reconnaissance des voyelles. </div>

<div style="text-align: justify"> Par ailleurs, un faible nombre de répétitions par rapport au nombre d'erreurs pour un stimulus donné suggère que quelques étudiant.e.s sont certain.e.s de reconnaître un stimulus alors que leur intuition est erronée. Plusieurs facteurs peuvent être mis en jeu : les sound files n'onts pas été enregistrés dans une chambre sourde et les accents peuvent influencer l'interprétation des stimuli. L'étude de ce phénomène constitue un axe intéressant de recherche.

Si l'on regarde les résultats de certain.e.s étudiant.e.s, nous pouvons observer ce même phénomène de manière plus tranchée puisqu'il y a moins d'occurrences.
