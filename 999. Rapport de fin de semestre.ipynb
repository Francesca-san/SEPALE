Mon stage au sein du projet SEPALE a débuté le 25 mai 2021. Durant les premières semaines nous nous sommes concentrés sur l’analyse des réponses au questionnaire soumis aux étudiant.e.s de L1, L2 et L3 de l’Université de Paris. Les réponses aux questions ont été discutées dans un rapport intitulé <ins> Rapport à l’issue du 1er semestre du projet SEPALE</ins>. 

Dans un second temps nous avons commencé l’analyse des logfiles, fichiers crées automatiquement lorsqu’un.e étudiant.e charge un exercice sur son ordinateur. Il existe quatre types de fichiers, les 2AFC, 5AFC, AX et Oddity qui correspondent à différents exercices de discrimination de voyelle. Afin d’analyser le contenu des fichiers la première étape est de les découper, de les « parser ».



*Figure 1*

Sur la figure ci-dessus (*Figure 1*), la première ligne correspond à attributs[0] = date_line.
La seconde à attributs[1] = keys.
Nous avons ensuite attributs[2:-1] = data.
Et enfin, attributs [-1] = stats_total.

Comme la première ligne l’indique, nous avons affaire à un fichier de type AFC2 qui propose de faire la différence entre les voyelles /ᴂ/ et /ʌ/. La ligne de keys labellise les informations qui sont en colonnes, les données (= data). Enfin, un exercice fini est clôturé par une dernière ligne qui correspond à ce que nous avons appelé « stats_total » et qui répertorie le nombre d’erreurs et le temps total.

J’ai mentionné précédemment le besoin de découper les fichiers. Ce besoin réside dans l’information que nous avons besoin de relever dans nos fonctions lorsque nous appelons des fonctions. Par exemple, lorsque nous utilisons la fonction criteria_by_key, il est essentiel de faire la différence entre les clés qui nous intéressent (« Sound File », « Stimulus », « Vowel ») et les critères ( = « criteria ») que nous allons confronter afin d’obtenir des résultats (« Response Time », « Repetitions », « NbErreurs »). 

## I. Organisation des notebooks: 

Les notebooks sont organisés dans un ordre hiérarchique. Les premiers correspondent à une « analyse » sommaire des fichiers AFC. Il s’agit-là d’une introduction à des fonctions très utiles, comme os.listdir(), qui crée une liste des éléments dans un fichier ; replace() qui permet de remplacer un élément par un autre (ex : replace("\n", " "), remplace les sauts de ligne par du vide) ; ou encore append(), qui permet d’ajouter à un ensemble vide (dictionnaire, liste) des éléments dans une boucle for. 

Le premier notebook 01_Read_and_clean_AFC_logfiles retranscrit les premiers contacts avec les fichiers AFC en rassemblant des appels des fonctions qui sont par la suite intégrés dans des scripts python pour traiter les fichiers. 

Le second fichier, 02_simple_AFC_file_analysis, rassemble des fonctions utiles si l’on veut avoir un aperçu des statistiques pour un seul exercice. En revanche, ces calculs ne fonctionnent que sur un simple logfile, ils ne peuvent traiter un grand nombre d’exercices à la fois ce qui compromet leur utilité. 

C’est à partir du fichier numéro 4, que l’on commence l’analyse de plusieurs fichiers de manière simultanée.
